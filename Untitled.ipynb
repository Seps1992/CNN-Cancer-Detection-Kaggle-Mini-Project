Step 1 - Introduction and Data Overview

This script addresses the binary classification challenge of detecting metastatic breast cancer in histopathologic images.
We work with small RGB patches (96Ã—96 pixels) extracted from whole-slide scans. The dataset consists of ~220,000 labeled TIFF images:
- Each image ID is mapped to a label (0 = no tumor, 1 = tumor) in a CSV.
- No duplicate samples; ~60% negative, ~40% positive.

We will perform EDA, build two distinct CNNs in PyTorch, compare their performance, and conclude with lessons learned.

```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve,
)
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator

RANDOM_SEED = 123
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

CSV_PATH   = "Downloads/kaggle/train_labels.csv"
IMG_FOLDER = "Downloads/kaggle/train"
BATCH_SIZE = 64
IMG_SIZE   = (32, 32)
```

Step 2 - EDA: load labels and inspect distribution

```python
# Load labels
labels_df = pd.read_csv(CSV_PATH)
print(f"Total samples: {len(labels_df)}")
print(labels_df["label"].value_counts())

# Plot class imbalance
counts = labels_df["label"].value_counts().sort_index()
plt.figure(figsize=(5,4))
plt.bar(["No Tumor", "Tumor"], counts.values, color=["skyblue","salmon"])
plt.title("Class Distribution")
plt.ylabel("Count")
plt.show()

# Display a handful of patches
sample_df = labels_df.sample(6, random_state=RANDOM_SEED)
fig, axes = plt.subplots(2, 3, figsize=(12,8))
for ax, (_, row) in zip(axes.flatten(), sample_df.iterrows()):
    img = np.array(Image.open(f"{IMG_FOLDER}/{row['id']}.tif"))
    ax.imshow(img)
    ax.set_title(f"Label={row['label']}")
    ax.axis("off")
plt.tight_layout()
plt.show()

# Pixel intensity histogram (flatten RGB channels)
pixels = np.vstack([
    np.array(Image.open(f"{IMG_FOLDER}/{rid}.tif")).reshape(-1,3)
    for rid in sample_df["id"]
])
plt.figure(figsize=(6,4))
plt.hist(pixels.ravel(), bins=50, alpha=0.6, color='gray')
plt.title("Pixel Intensity Distribution")
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.show()

def center_crop(img: np.ndarray) -> np.ndarray:
    """Crop the 96Ã—96 array to its central 32Ã—32 region."""
    h, w = img.shape[0], img.shape[1]
    cy, cx = h // 2, w // 2
    off = 16
    return img[cy-off:cy+off, cx-off:cx+off, :]
```

Step 3 - Data Generators

```python
datagen = ImageDataGenerator(
    preprocessing_function=lambda im: center_crop(im),
    rescale=1/255,
    rotation_range=15,
    horizontal_flip=True,
    validation_split=0.2
)

# Add filename column
labels_df["filename"] = labels_df["id"] + ".tif"
labels_df["label_str"] = labels_df["label"].astype(str)

train_gen = datagen.flow_from_dataframe(
    labels_df,
    directory=IMG_FOLDER,
    x_col="filename",
    y_col="label_str",
    subset="training",
    batch_size=BATCH_SIZE,
    target_size=IMG_SIZE,
    class_mode="binary",
    seed=RANDOM_SEED
)

val_gen = datagen.flow_from_dataframe(
    labels_df,
    directory=IMG_FOLDER,
    x_col="filename",
    y_col="label_str",
    subset="validation",
    batch_size=BATCH_SIZE,
    target_size=IMG_SIZE,
    class_mode="binary",
    shuffle=False,
    seed=RANDOM_SEED
)
```

Step 4 - Model Architectures

```python
def build_base_cnn(input_shape=(32,32,3)):
    m = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32,3,padding='same',activation='relu'),
        layers.MaxPool2D(),
        layers.Conv2D(64,3,padding='same',activation='relu'),
        layers.MaxPool2D(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1, activation='sigmoid', dtype='float32')
    ])
    return m

def build_bn_cnn(input_shape=(32,32,3)):
    m = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(32,3,padding='same'),
        layers.BatchNormalization(), layers.Activation('relu'),
        layers.MaxPool2D(),
        layers.Conv2D(64,3,padding='same'),
        layers.BatchNormalization(), layers.Activation('relu'),
        layers.MaxPool2D(),
        layers.Flatten(),
        layers.Dense(128),
        layers.BatchNormalization(), layers.Activation('relu'),
        layers.Dropout(0.4),
        layers.Dense(1, activation='sigmoid', dtype='float32')
    ])
    return m

# Use MirroredStrategy for multi-GPU / mixed precision if available
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)

    # Use NCCL all-reduce for GPU-to-GPU communication
    cross_device = tf.distribute.NcclAllReduce()
    strategy = tf.distribute.MirroredStrategy(cross_device_ops=cross_device)

    # Mixed precision still applies
    tf.keras.mixed_precision.set_global_policy('mixed_float16')
    print(f"Using {strategy.num_replicas_in_sync} GPUs with NCCL all-reduce")
else:
    strategy = tf.distribute.OneDeviceStrategy("/cpu:0")
    print("No GPU found, using CPU")
```

Step 5 - Hyperparameter Sweep

```python
def trial_run(build_fn, lr, epochs=5):
    with strategy.scope():
        model = build_fn()
        model.compile(
            optimizer=tf.keras.optimizers.Adam(lr),
            loss='binary_crossentropy',
            metrics=['accuracy',
                     tf.keras.metrics.AUC(name='val_auc')]
        )
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=epochs,
        verbose=2
    )
    return history, model

learning_rates = [1e-3, 5e-4]
hist_base = {}
for lr in learning_rates:
    print(f"\n=== BaseCNN @ lr={lr} ===")
    hist_base[lr], _ = trial_run(build_base_cnn, lr)

hist_bn = {}
for lr in learning_rates:
    print(f"\n=== BNCNN @ lr={lr} ===")
    hist_bn[lr], _ = trial_run(build_bn_cnn, lr)

# Plot comparison of validation AUC
plt.figure(figsize=(6,4))
for lr, h in hist_base.items():
    plt.plot(h.history['val_auc'], label=f"Base@{lr}")
for lr, h in hist_bn.items():
    plt.plot(h.history['val_auc'], '--', label=f"BatchNorm@{lr}")
plt.title("Validation AUC")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.legend()
plt.show()
```

Step 6: Confusion Matrix & Classification Report

```python
# Retrain best model (BatchNorm with lr=5e-4) for final evaluation
_, best_model = trial_run(build_bn_cnn, 5e-4, epochs=10)

# Predictions on validation set
val_gen.reset()
y_prob = best_model.predict(val_gen, verbose=0).ravel()
y_pred = (y_prob > 0.5).astype(int)
y_true = val_gen.labels

# Metrics
print(classification_report(y_true, y_pred, digits=4))
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(4,4))
plt.imshow(cm, cmap='Blues')
plt.colorbar()
plt.xticks([0,1], ['No Tumor','Tumor'])
plt.yticks([0,1], ['No Tumor','Tumor'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
for i in [0,1]:
    for j in [0,1]:
        plt.text(j, i, cm[i,j], ha='center', va='center', color='white')
plt.show()

fpr, tpr, _ = roc_curve(y_true, y_prob)
plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, label=f"AUC={roc_auc_score(y_true,y_prob):.4f}")
plt.plot([0,1],[0,1],'--',color='gray')
plt.title("ROC Curve")
plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend()
plt.show()
```

Step 8: Prepare Test Predictions and Submission

```python
# 1) Build a DataFrame of test IDs & filenames
test_dir = "Downloads/kaggle/test"
test_files = [f for f in os.listdir(test_dir) if f.endswith(".tif")]
test_df = pd.DataFrame({
    "id":       [os.path.splitext(f)[0] for f in test_files],
    "filename": test_files
})

# 2) Create a test generator using the same center-crop + rescale
test_datagen = ImageDataGenerator(
    preprocessing_function=lambda im: center_crop(im),
    rescale=1/255
)
test_gen = test_datagen.flow_from_dataframe(
    test_df,
    directory=test_dir,
    x_col="filename",    # MUST match the column we created above
    y_col=None,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode=None,
    shuffle=False
)

# 3) Predict probabilities on the test set
print("Predicting on test setâ€¦")
probs = best_model.predict(test_gen, verbose=1).ravel()

# 4) Threshold at 0.5 to obtain final labels
preds = (probs > 0.5).astype(int)

# 5) Build submission DataFrame and save to CSV
submission = pd.DataFrame({
    "id":    test_df["id"],
    "label": preds
})
submission.to_csv("submission.csv", index=False)
print("Submission saved to submission.csv:")
print(submission.head())
```

Step 7: Conclusion
BatchNorm CNN converged faster and achieved ~3% higher AUC vs the baseline. 
Lower learning rate (5e-4) was more stable than 1e-3. 
Future directions: transfer learning, stronger augmentations (mixup/cutmix), 
and explainable AI overlays to highlight regions driving the predictions.
